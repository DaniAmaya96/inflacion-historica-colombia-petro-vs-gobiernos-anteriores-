```python
%%manim -qm -v WARNING Sum_of_deviations

from manim import *
import numpy as np

class Sum_of_deviations(Scene):
    def construct(self):
        # 1. Configuración de Cámara
        self.camera.background_color = "#111111" # Gris muy oscuro (casi negro)

        # ---------------------------------------------------------
        # PARTE 1: TEORÍA Y ÁLGEBRA (Tu código original integrado)
        # ---------------------------------------------------------

        # --- PORTADA ---
        Title = Text("Why is the sum of the deviations from the mean equal to zero?", font_size=30, color=BLUE)
        Author = Text("Black and Wolf Productions", font_size=25, color=PINK).next_to(Title, DOWN)
        self.play(Write(Title))
        self.wait(1)
        self.play(Write(Author))
        self.wait(1)
        self.clear()

        # --- DEFINICIÓN MEDIA ---
        Texto1 = Text("The statistical Mean can be defined as:", font_size=25, color=WHITE).to_edge(UP)
        Mean = MathTex(r"\mu = \frac{1}{n}\sum_{i=1}^{n}x_i")
        self.play(Write(Texto1))
        self.play(Write(Mean))
        self.wait(1)
        self.play(FadeOut(Texto1), FadeOut(Mean))

        # --- DEFINICIÓN DESVIACIÓN ---
        Texto2 = Text("The Deviation from an observation to the mean:", font_size=25, color=WHITE).to_edge(UP)
        Deviation = MathTex(r"d_i = x_i - \mu")
        self.play(Write(Texto2))
        self.play(Write(Deviation))
        self.wait(1)
        self.play(FadeOut(Texto2), FadeOut(Deviation))

        # --- LA PREGUNTA ---
        Texto3 = Text("Then the sum of the deviations from the mean is:", font_size=25, color=WHITE).to_edge(UP)
        Sum = MathTex(r"\sum_{i=1}^{n}(x_i-\mu)")
        WhantShow = MathTex(r"\sum_{i=1}^{n}(x_i-\mu)= 0")
        Texto4 = Text("which is equal to zero", font_size=25, color=YELLOW).next_to(WhantShow, DOWN)

        self.play(Write(Texto3))
        self.play(Write(Sum))
        self.wait(1)
        self.play(ReplacementTransform(Sum, WhantShow))
        self.play(Write(Texto4))
        self.wait(1)
        self.play(FadeOut(Texto3), FadeOut(WhantShow), FadeOut(Texto4))

        # --- DEMOSTRACIÓN ALGEBRAICA ---
        Texto5 = Text("Let's understand why:", font_size=25, color=WHITE).to_edge(UP)

        # Ecuaciones paso a paso
        eq1 = MathTex(r"\sum_{i=1}^{n}(x_i-\mu)= 0")
        eq2 = MathTex(r"\sum_{i=1}^{n}x_i - \sum_{i=1}^{n} \mu")
        eq3 = MathTex(r"\sum_{i=1}^{n}x_i - n\mu")
        eq4 = MathTex(r"\sum_{i=1}^{n}x_i - n \frac{1}{n}\sum_{i=1}^{n}x_i")
        eq5 = MathTex(r"\sum_{i=1}^{n}x_i - \frac{n}{n}\sum_{i=1}^{n}x_i")
        eq6 = MathTex(r"\sum_{i=1}^{n}x_i - \sum_{i=1}^{n}x_i = 0")
        eq7 = MathTex(r"0").set_color(PINK)

        self.play(Write(Texto5))
        self.play(Write(eq1))
        self.wait(1)
        self.play(ReplacementTransform(eq1, eq2))
        self.wait(1)
        self.play(ReplacementTransform(eq2, eq3))
        self.wait(1)
        self.play(ReplacementTransform(eq3, eq4))
        self.wait(1)
        self.play(ReplacementTransform(eq4, eq5))
        self.wait(1)
        self.play(ReplacementTransform(eq5, eq6))
        self.wait(1)
        self.play(ReplacementTransform(eq6, eq7))
        self.wait(2)
        self.clear()

        # ---------------------------------------------------------
        # PARTE 2: SIMULACIÓN NUMÉRICA (TABLA VISIBLE)
        # ---------------------------------------------------------

        # Título Sección
        txt_simulacion = Text("Numerical Simulation", font_size=30, color=BLUE).to_edge(UP)
        self.play(Write(txt_simulacion))
        self.wait(1)

        # 1. Generar Datos "Amigables" (Media Entera)
        np.random.seed(50) # Cambié la semilla para variar
        n_obs = 6

        # Generamos 5 números aleatorios
        data_parcial = np.random.randint(1, 8, size=n_obs - 1)

        # Calculamos cuál debe ser el último número para que la media sea entera
        # Truco: La suma total debe ser divisible por n_obs (6)
        suma_actual = np.sum(data_parcial)
        resto = suma_actual % n_obs

        # El último número es lo que falta para el siguiente múltiplo de 6
        ultimo_numero = (n_obs - resto) if resto != 0 else 0

        # A veces el último número queda muy chico o cero, le sumamos 6 para que tenga "cuerpo"
        while ultimo_numero < 1:
            ultimo_numero += n_obs

        # Unimos todo
        data = np.append(data_parcial, ultimo_numero)

        # Ahora sí, cálculos perfectos
        media = np.mean(data) # Esto dará algo como 3.0, 4.0, etc.
        media = int(media)    # Lo convertimos a entero para que se vea bonito (3 en vez de 3.0)
        deviation = data - media

        # 2. CREACIÓN MANUAL DE OBJETOS (Para asegurar visibilidad)
        # Creamos una lista donde guardamos los números como objetos MathTex BLANCOS
        filas_mobjects = []

        for i in range(n_obs):
            # i+1 es el índice, data[i] es el valor, deviation[i] es el resultado
            obj_idx = MathTex(str(i + 1), color=WHITE)
            obj_val = MathTex(str(data[i]), color=WHITE)
            obj_dev = MathTex(f"{deviation[i]:.1f}", color=WHITE)

            # Guardamos la fila completa en nuestra lista
            filas_mobjects.append([obj_idx, obj_val, obj_dev])

        # Creamos la fila de Totales aparte
        total_label = Text("Total", font_size=24, color=WHITE)
        total_val = MathTex(str(np.sum(data)), color=WHITE)
        total_dev = MathTex(str(round(np.sum(deviation))), color=RED) # Rojo para resaltar

        # La agregamos al final de nuestra lista
        filas_mobjects.append([total_label, total_val, total_dev])

        # 3. CREAR LA TABLA (Solo se usa para organizar la posición)
        tabla = MobjectTable(
            filas_mobjects,
            col_labels=[
                MathTex("i", color=YELLOW),
                MathTex("x_i", color=YELLOW),
                MathTex(r"x_i - \mu", color=YELLOW)
            ],
            include_outer_lines=True,
            line_config={"stroke_width": 1, "color": WHITE}
        ).scale(0.5)

        # Texto de la media
        txt_media = MathTex(fr"\mu = {media:.1f}", color=YELLOW).next_to(tabla, UP)

        # --- ANIMACIÓN ---

        # A. Mostrar la rejilla y encabezados
        self.play(
            Create(tabla.get_horizontal_lines()),
            Create(tabla.get_vertical_lines()),
            Write(tabla.get_col_labels())
        )

        # B. Llenar columnas 1 y 2 (Índice y Valor)
        # IMPORTANTE: Animamos los objetos de NUESTRA lista, no pedimos celdas a la tabla
        for i in range(n_obs):
            mi_idx = filas_mobjects[i][0]
            mi_val = filas_mobjects[i][1]
            self.play(Write(mi_idx), Write(mi_val), run_time=0.2)

        # C. Mostrar Media
        self.play(Write(txt_media))

        # D. Calcular Desviaciones (Columna 3)
        for i in range(n_obs):
            mi_dev = filas_mobjects[i][2]

            # Usamos la tabla SOLO para saber dónde dibujar el rectángulo azul
            # get_rows()[i+1] porque la fila 0 son los encabezados
            rect = SurroundingRectangle(tabla.get_rows()[i+1], color=BLUE, buff=0.1)

            self.play(Create(rect), run_time=0.2)
            self.play(Write(mi_dev)) # Animamos el número visible
            self.play(FadeOut(rect), run_time=0.1)

        # E. Totales y Gran Final
        fila_final = filas_mobjects[-1] # La última fila de nuestra lista

        # Escribimos "Total" y la suma de datos
        self.play(Write(fila_final[0]), Write(fila_final[1]))
        self.wait(0.5)

        # Animación especial para el CERO
        cero_final = fila_final[2]

        # Transformación visual desde el título de la columna hasta el cero
        self.play(TransformFromCopy(tabla.get_col_labels()[2], cero_final))

        # Recuadro rojo final
        box_final = SurroundingRectangle(cero_final, color=RED)
        self.play(Create(box_final))

        self.wait(3)
```
